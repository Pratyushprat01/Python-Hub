```python
apple.rename(columns = {
    'iPhone Sales (in million units)':'iPhone_Sales', 'iPad Sales (in million units)':'iPad_Sales',
    'Mac Sales (in million units)': 'Mac_Sales', 'Wearables (in million units)':'Wearables',
    'Services Revenue (in billion $)':'Revenue'}, inplace = True)
apple
```

```python
apple.columns = apple.columns.str.strip()
```

```python
# Total sales by Region
region_summary = apple.groupby('Region').agg({
    'iPhone_Sales': 'sum',
    'iPad_Sales': 'sum',
    'Mac_Sales': 'sum',
    'Wearables': 'sum',
    'Revenue': 'sum'
}).reset_index()

print(region_summary)
```

```python
# Sales contribution bt product
apple['Total Sales'] = apple['iPhone_Sales']+ apple['Mac_Sales']+ apple['Wearables']+apple['iPad_Sales']
product_contribution = apple[['iPhone_Sales', 'iPad_Sales', 'Mac_Sales', 'Wearables']]
product_contribution_percentage  = (product_contribution/ product_contribution.sum())*100
print(product_contribution_percentage)
```

```python
# The top performing state
top_state = apple.groupby('State')['Revenue'].sum().sort_values(ascending = 5).head(5)
top_state
```

```python
region_summary.plot(kind = 'bar', x ='Region', y = ['iPhone_Sales', 'iPad_Sales', 'Mac_Sales', 'Wearables' ])
plt.title('Total Sales by Region')
plt.ylabel('Sales(In million Units)')
plt.show()
```

```python
# Pie chart : Product_contrbution

product_contribution.plot(kind = 'pie', autopct = '%1.1f%%',figsize=(5,5), labels = product_contribution.index)
plt.ylable('')
plt.show()
```

```python
top_state.plot(kind = 'bar', figsize = (5,5), color = "Violet")
plt.title('Top 5 States by Revenue')
plt.ylabel('Revenue (in Billion $)')
plt.xlabel('State')
plt.show()
```

```python
# Save region summary and product contribution to CSV
region_summary.to_csv('region_summary.csv', index=False)
product_contribution_percentage.to_csv('product_contribution.csv', index=False)
```
